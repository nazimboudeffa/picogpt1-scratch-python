# PicoGPT — Test du modèle
Ce notebook charge le modèle entraîné et génère une réponse à partir d'un prompt.

## 1. Chargement du modèle

```python
from transformers import pipeline

pipe = pipeline(
    "text-generation",
    model="./PicoGPT1-Scratch",
    tokenizer="./PicoGPT1-Scratch",
)

print("Modèle chargé !")
```

## 2. Génération de texte
Modifie le `prompt` ci-dessous et relance la cellule pour tester différentes entrées.

```python
prompt = "What is the capital of France?\n"
max_new_tokens = 50

out = pipe(prompt, max_new_tokens=max_new_tokens, do_sample=False)
print(out[0]["generated_text"])
```
